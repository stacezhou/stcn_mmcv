```python
with open('breakdance_all.pkl','rb') as f:
    output = pickle.load(f)
```

```python
def vis(output,name,metric='J',length = 120):
    video_result = output[name]
    if metric == 'J' :
        m = 0
    else:
        m = 1
    result_JF = np.zeros((2,42,length))
    for frame_index,(J,F) in video_result:
        frame_index = frame_index[1:]
        J=J[1:]
        F=F[1:]
        gt_index = [x-frame_index[0]+21 for x in frame_index]
        result_JF[0][gt_index,frame_index] = J
        result_JF[1][gt_index,frame_index] = F

    plt.figure(figsize=(20,10))
    data = result_JF[m]
    mask = np.zeros_like(data)
    mask[data == 0] = True
    # choose 5 colours to create 5 bins
    fig = sns.heatmap(data,mask=mask,cmap="YlGnBu",vmin=0.6,vmax=1)
    fig.set_yticklabels([x if x % 3 == 0 else '' for x in range(-21,41-20)])
    # fig.set_xticklabels([x if x % 5 == 0 else '' for x in range()])
    plt.ylabel('start frame location')
    plt.xlabel('frame id')
    plt.title(f'{metric} score {name}')
    plt.show()
```

```python
vis(output,'india',length=85)
```
